## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variabllaunchSite FROM SPACEXTBL   """
query = """ SELECT DISTINCT LaunchSite FROM  SPACEXTBL  """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variabllaunchSite FROM SPACEXTBL   """
query = """ SELECT  LaunchSite, 
FlightNumber,
BoosterVersion
FROM  SPACEXTBL 
WHERE 
LaunchSite like("CCS%") """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
#Rellena la variabllaunchSite FROM SPACEXTBL   """
query = """ SELECT
PayloadMass,
Customers  
FROM  SPACEXTBL 
WHERE 
customers = "NASA (CRS)" """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """ SELECT 
avg(PayloadMass),
Serial 
FROM SPACEXTBL 
WHERE
Serial like("B1%")
 group by Serial """

df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """ SELECT 
Date,
LandingOutcome
FROM SPACEXTBL
WHERE
LandingOutcome = "TRUE RTLS" """

df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ SELECT 
LandingOutcome,
PayloadMass,
BoosterVersion
FROM SPACEXTBL
WHERE
LandingOutcome = "TRUE RTLS" and
PayloadMass between 400 AND 6000
"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """ 
SELECT  

count(case when LandingOutcome = "TRUE ASDS" or LandingOutcome = "True RTLS" then FlightNumber end) as "numero_de_aciertos",
count(case when LandingOutcome = "False Ocean" then FlightNumber end) as "numero_de_fallos"
FROM SPACEXTBL

"""
df  = pd.read_sql(query, con = engine )
df
```
